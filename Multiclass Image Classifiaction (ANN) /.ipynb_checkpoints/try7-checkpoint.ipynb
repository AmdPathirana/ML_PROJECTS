```python
#import tensorflow
import tensorflow
```

```python
#import data set 
from tensorflow.keras.datasets import fashion_mnist
data = fashion_mnist.load_data()
```

```python
#assgined data to train and text 
train, test = data
```

```python
#split data into train and text 
x_train, y_train = train
x_test,  y_test = test
```

```python
#check the shape of the data 
print("x_train: {}".format(x_train.shape))
print("x_test: {}".format(x_test.shape))
print("y_train: {}".format(y_train.shape))
print("x_test: {}".format(y_test.shape))
```

```python
#normalized the data 
x_train = x_train/255
x_test = x_test/255
```

```python
#check the row data values 
x_train[34]
```

```python
#check the corosponding the label 
import matplotlib.pyplot as plt 

z=192 

plt.imshow(x_train[z])


#get the name 
class_names=['0 T-shirt/top', '1 Trouser', '2 Pullover', '3 Dress', '4 Coat',
    '5 Sandal', '6 Shirt', '7 Sneaker', '8 Bag', '9 Ankle boot']

def get_name(n):
    return class_names[n]

print(get_name(y_train[z]))

```

```python
#image reshape to vector (matrix to vector) 
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
```

```python
#check the shape
x_test.shape
```

```python
x_train[3].shape
```

```python
#create the model 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

from tensorflow.keras.regularizers import l1_l2
l1_reg = 0.001
l2_reg = 0.001

model = Sequential()

model.add(Dense(256, input_shape=(784,), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(56, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
#model compile
model.compile(loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'], optimizer='adam')
```

```python
#early stoping 
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=10, verbose=1)
#model fit 
history = model.fit(x_train, y_train, epochs=80, validation_data=(x_test, y_test), verbose=1, batch_size=20, callbacks=(early_stop) )
```

```python
train_accuracy = history.history['sparse_categorical_accuracy']
test_accuracy = history.history['val_sparse_categorical_accuracy']

train_loss = history.history['loss']
test_loss  = history.history['val_loss']


epoch = range(1, len(train_accuracy)+1)

plt.figure(figsize=(8,5))
plt.plot(train_accuracy, epoch, 'b', label='Train Accuracy')
plt.plot(test_accuracy, epoch, 'r', label='Test Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()


plt.figure(figsize=(8,5))
plt.plot(train_loss, epoch, 'b', label='Train Loss')
plt.plot(test_loss, epoch, 'r', label='Test Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
#model evaluation 
loss, accuracy = model.evaluate(x_test, y_test)

print("Loss: {}".format(loss))
print("Accuracy: {}".format())
```

```python
#get the model predicton 
predictions = model.predict(x_test)

#get the max value of the predictoin list using argmax function 
import numpy as np

prediction_values = np.argmax(predictions, axis=1)
```

```python
#check the predciton probabilty list and shape
predictions.shape
```

```python
#confusion matrix 

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(prediction_values, y_test)
cm
```

```python
#use heatmap to get the final confusion matrix 

import seaborn as sns

sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
```
